#### Now let's work with the "south belt" of madrid. Make a subsetting that contains the following populations (level5): "Fuenlabrada","Leganés","Getafe","Alcorcón" (★★☆) 

```python
#Utilizo OR para seleccionar todas las poblaciones 
ds_south_belt =  ds.loc[(ds["level5"] == "Fuenlabrada") | (ds["level5"] == "Leganés") | (ds["level5"] == "Getafe") | (ds["level5"] == "Alcorcón")]

ds_south_belt
```

#### Make a bar plot of the median of prices (★★★) 

```python
#Agrupo los precios por locaclidad, calculo la mediana y reseteo el indice para que me agrupe los datos
median_prices = ds_south_belt.groupby('level5')['price'].median().reset_index()

# Creo un gráfico de barras de la mediana de precios
plt.bar(median_prices['level5'], median_prices['price'],edgecolor = 'black')
plt.xlabel('Localidad')
plt.ylabel('Mediana de Precios')
plt.title('Mediana por localidad')
plt.show()
```

#### Normalize the variable of prices for each population and plot the 4 histograms in the same plot (★★★) 

```python
#Hago una copia para trabajar sin errores
normalizar_y_mostrar = ds_south_belt.copy()

#Agrupo por localidad
localidades = normalizar_y_mostrar.groupby('level5')

#Normalizo la variable de precios con una funcion

def normalizar(x):
    return (x - x.min()) / (x.max() - x.min())

normalizar_y_mostrar['precio_norm'] = localidades['price'].transform(normalizar)


Fuenlabrada_hist = localidades.get_group('Fuenlabrada')
Leganes_hist = localidades.get_group('Leganés')
Getafe_hist = localidades.get_group('Getafe')
Alcorcon_hist = localidades.get_group('Alcorcón')

#Muestro el histograma con los precios normalizados
plt.figure(figsize=(10, 8))

# Histograma 1
plt.subplot(2, 2, 1)
plt.hist(Fuenlabrada_hist['precio_norm'], bins=50, edgecolor='k')
plt.title('Fuenlabrada')

# Histograma 2
plt.subplot(2, 2, 2)
plt.hist(Leganes_hist['precio_norm'], bins=50, edgecolor='k')
plt.title('Leganés')

# Histograma 3
plt.subplot(2, 2, 3)
plt.hist(Getafe_hist['precio_norm'], bins=50, edgecolor='k')
plt.title('Getafe')

# Histograma 4
plt.subplot(2, 2, 4)
plt.hist(Alcorcon_hist['precio_norm'], bins=50, edgecolor='k')
plt.title('Alcorcón')

plt.tight_layout()  # Ajustar automáticamente los espacios entre los subgráficos
plt.show()
```

Histograma con los precios normalizados

#### What can you say about the price per square meter between the towns of Getafe and Alcorcón?

```python
#Almaceno los nombres de las ciudades a consultar
ciudad1 = 'Getafe'
ciudad2 = 'Alcorcón'

#Consigo los datos de cada ciudad
data_ciudad1 = ds.loc[ds['level5'] == ciudad1].copy()
data_ciudad2 = ds.loc[ds['level5'] == ciudad2].copy()

#Calculo el precio por metro cuadrado
data_ciudad1['precio_metro_cuadrado'] = data_ciudad1['price'] / data_ciudad1['surface']
data_ciudad2['precio_metro_cuadrado'] = data_ciudad2['price'] / data_ciudad2['surface']

#Calculo la media de cada uno
media_ciudad1 = data_ciudad1['precio_metro_cuadrado'].mean()
media_ciudad2 = data_ciudad2['precio_metro_cuadrado'].mean()

print(f'El precio promedio de {ciudad1} por metro cuadrado es de: {media_ciudad1}')
print(f'El precio promedio de {ciudad2} por metro cuadrado es de: {media_ciudad2}')
```

El precio promedio por metro cuadrado en Alcorcon es 173 euros aproximadamente mas caro que en Getafe
